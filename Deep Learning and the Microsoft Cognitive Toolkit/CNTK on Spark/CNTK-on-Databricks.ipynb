#CNTK init script
Run this notebook to setup an init script to install CNTK on a CPU or GPU-enabled cluster:
1. Import this notebook into your workspace
2. Edit the version of CNTK you want to install, pick the compute device and set the name of the cluster. 
3. Run this notebook (clicking ``Run All`` above).

The init script will install OpenMPI on all the nodes and the Python 2.7 bindings for CNTK.

```
cntkVersion = "2.0.beta15.0"
useGPU = False
clusterName = "cntk"
```

You do not need to edit the rest of this notebook

```
device = "GPU" if useGPU else "CPU-only"
```

```
script = """
#!/usr/bin/env bash
CNTK_VER="{cntkVersion}"
CNTK_URL="https://cntk.ai/PythonWheel/{device}/cntk-$CNTK_VER-cp27-cp27mu-linux_x86_64.whl"
echo "Running apt-get upgrade"
apt-get upgrade

echo "Installing Open MPI"
apt-get -y install openmpi-bin

echo "Installing CNTK via pip"
pip install --upgrade $CNTK_URL
""".format(cntkVersion=cntkVersion, device=device)
```

```
dbutils.fs.put("dbfs:/databricks/init/{clusterName}/install_cntk.sh".format(clusterName=clusterName), script, True)
```

```
print "Init scripts installed for cluster: " + clusterName
print("\n".join([x.name for x in dbutils.fs.ls("dbfs:/databricks/init/%s" % clusterName)]))
```
